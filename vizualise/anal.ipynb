```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
```

```python
from functions import split_job, count_column
```

```python
files = ['data/business_analyst.csv', 'data/data_analyst.csv', 'data/financial_analyst.csv']
```

```python
df = pd.DataFrame()
for f in files:
    print('___-=',f)
    d = pd.read_csv(filepath_or_buffer=f, sep=',')
    df = df.append(d)
df = df.drop(columns=['query'])
print(df.shape)
df = df.drop_duplicates(keep='first')
print(df.shape)
```

```python
df1 = split_job(df, column='job',company='company1')
df2 = split_job(df, column='currentJob',company='company_now')
df3 = split_job(df, column='pastJob',company='company_past')
display(df2.tail(2))
df4 = pd.concat([df, df1,df2['company_now'],df3['company_past']], axis=1)
df4['company']= df4['company1'].fillna(df4['company_now']).fillna(df4['company_past'])
```
